## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

## Loading and reading the dataset

```python
iris = pd.read_csv("Iris.csv")
iris.drop(["Id"], axis = 1, inplace = True)
iris.head()
```

## Basic data exploration

```python
# basic info about the data
iris.info()
```

```python
# summary of data
iris.describe()
```

```python
# gives the number of rows and columns in dataset
iris.shape
```

```python
#shows the count of the iris species
iris['Species'].value_counts()
```

## Data Preparation

```python
# features- predictor
features  = list(iris.columns)[:-1]
print(features)
# predicted
target = list(iris.columns)[-1:][0]
print(target)

# we will use features to predict the kind of iris species.
```

```python
# Separates features and corresponding labels/target 
# by dropping Species - we get data frame with all features 

X = iris.drop(['Species'], axis=1)  #  X - features
Y = iris['Species'] # Y - target

print(X.shape) #dimensions of input data
print(Y.shape) #dimensions of output data
```

## Dividing data into training and testing data

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size = 0.25, random_state = 0) 
print(X_train.shape)
print(X_test.shape)
```

```python
#Training using Decision Tree Classifier 
#from sklearn.tree import DecisionTreeClassifier  
#dtree = DecisionTreeClassifier(criterion='gini')  
#dtree.fit(X_train, Y_train) 
#print(dtree)

#ypred1 = dtree.predict(X_test)  
#print(ypred1)

#acc_1 = accuracy_score(Y_test,ypred1)
#print("Accuracy for Gini model {} %".format(acc_1*100))
```

```python
dtree=DecisionTreeClassifier(criterion = "entropy")
dtree.fit(X_train,Y_train)
```

```python
# predict using classifier 

Y_pred = dtree.predict(X_test)  
print(Y_pred)
```

```python
# computing accuracy 
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix   

accuracy = accuracy_score(Y_test,Y_pred)
print("Accuracy for Entropy model {} %".format(accuracy*100))
```

```python
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred)) 
```

```python
#Plotting the decision tree

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)

tree.plot_tree(dtree, feature_names = ['SepalLen','SepalWid','PetalLen', 'PetalWid'],
               class_names = 'Species',filled = True,rounded = True, proportion = False, precision = 2);

fig.savefig('imagename.png')
```

```python
# Feeding the new test data for classifier to predict correctly
print(dtree.predict([[3,2,1,5]]))
print(dtree.predict([[5,3,1,0]]))
print(dtree.predict([[5,2,3,1]]))
```
